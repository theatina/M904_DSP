```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import sounddevice as sd
```

```python
t = np.arange(44100)/44100
```

```python
f = 333

s1 = np.sin( 2*np.pi*f*t )

s_square = (s1>0)*1 + (s1<0)*(-1)

```

```python
fig, plots = plt.subplots(1,1,figsize=(15,3))
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

plots.grid(True)
plots.title.set_text(r"$\bf{s_{11} = sin( 2πf_1t )  }$")
plots.plot( s_square[:2205] )
```

```python
f = 400

s_sawtooth = t - np.floor(t)
```

```python
fig, plots = plt.subplots(1,1,figsize=(15,3))
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

plots.grid(True)
plots.title.set_text(r"$\bf{sawtooth function}$")
plots.plot( s_sawtooth[:2205] )
```
