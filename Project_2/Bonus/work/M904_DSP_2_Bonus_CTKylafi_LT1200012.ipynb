```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import sounddevice as sd
```

```python
s_r = 44100
t = np.arange(s_r)/s_r
```

```python
f_1 = 220 #A3
f_2 = 293.66 #D4

steps_to_plot = int(s_r/25)
print("\nTime steps to plot: %d"%(steps_to_plot))

s_1_1 = np.sin( 2*np.pi*f_1*t ) 
s_1_2 = np.sin( 2*np.pi*f_2*t ) 

s_1 = s_1_1 + s_1_2

fig, plots = plt.subplots(4,1,figsize=(15,10), sharex=True)
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

plots[0].grid(True)
plots[0].title.set_text(r"$\bf{s_{11} = sin( 2πf_1t )  }$")
plots[0].plot( s_1_1[:steps_to_plot], "|"  )

plots[1].grid(True)
plots[1].title.set_text(r"$\bf{s_{12} = sin( 2πf_2t )  }$")
plots[1].plot( s_1_2[:steps_to_plot] )
plots[1].plot( s_1_2[:steps_to_plot] )
plots[1].plot( s_1_2[:steps_to_plot] )
plots[1].plot( s_1_2[:steps_to_plot], "|" )

plots[2].grid(True)
plots[2].title.set_text(r"$\bf{ s_{11}\ &\ s_{12} }$")
plots[2].plot( s_1_1[:steps_to_plot], "|"  )
plots[2].plot( s_1_2[:steps_to_plot] )
plots[2].plot( s_1_2[:steps_to_plot] )
plots[2].plot( s_1_2[:steps_to_plot], "|"  )

plots[3].grid(True)
plots[3].title.set_text(r"$\bf{s_1 = s_{11} + s_{12} }$")
plots[3].plot( s_1[:steps_to_plot] )
plots[3].plot( s_1[:steps_to_plot] )
plots[3].plot( s_1[:steps_to_plot] )
plots[3].plot( s_1[:steps_to_plot] )
plots[3].plot( s_1[:steps_to_plot], "|" )

fig.suptitle(r"$\bf{Figure\ 1:\ s_1 = s_{11} + s_{12}}$")

# sd.play( s_1_1 , 44100*2 )
# sd.wait( 2 )
# sd.play( s_1_2 , 44100*2 )
# sd.wait( 2 )
# sd.play( s_1 , 44100*2 )
```
