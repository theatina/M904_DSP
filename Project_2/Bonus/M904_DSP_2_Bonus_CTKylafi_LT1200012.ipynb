```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import sounddevice as sd
```

```python
t = np.arange(44100)/44100
```

```python
f_1 = 200
f_2 = 300


s_1_1 = np.sin( 2*np.pi*f_1*t ) 
s_1_2 = np.sin( 2*np.pi*f_2*t )

s_1 = s_1_1 + s_1_2

fig, plots = plt.subplots(4,1,figsize=(12,12), sharex=True)
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

plots[0].grid(True)
plots[0].title.set_text(r"$\bf{s_{11} = sin( 2πf_1t )  }$")
plots[0].plot( s_1_1[:2205] )

plots[1].grid(True)
plots[1].title.set_text(r"$\bf{s_{12} = sin( 2πf_2t )  }$")
plots[1].plot( s_1_2[:2205] )

plots[2].grid(True)
plots[2].title.set_text(r"$\bf{s_1 = s_{11} + s_{12} }$")
plots[2].plot( s_1_1[:2205] )
plots[2].plot( s_1_2[:2205] )

plots[3].grid(True)
plots[3].title.set_text(r"$\bf{s_1 = s_{11} + s_{12} }$")
plots[3].plot( s_1[:2205] )



fig.suptitle(r"$\bf{Figure\ :\ s_1 = s_{11} + s_{12}}$")


# sd.play( s , 44100 )
```

```python
f_1 = 200
f_2 = 300

s_2_1_1 = np.sin( 2*np.pi*f_1*t )

s_2_2_1 = np.sin( 2*np.pi*f_2*t +  np.pi/4 )
s_2_1 = s_2_1_1 + s_2_2_1

s_2_2_2 = np.sin( 2*np.pi*f_2*t +  np.pi/2 )
s_2_2 = s_2_1_1 + s_2_2_2

s_2_2_3 = np.sin( 2*np.pi*f_2*t +  np.pi )
s_2_3 = s_2_1_1 + s_2_2_3

s_2_2_4 = np.sin( 2*np.pi*f_2*t +  3*np.pi/2 )
s_2_4 = s_2_1_1 + s_2_2_4

fig = plt.figure(figsize=(15,3))
plt_s2_1 = fig.add_subplot(1, 1, 1)

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title(r"$\bf{Figure\ :\ s_{211}  = sin( 2π*f_1*t )}$")
plt.gcf().subplots_adjust(bottom=0.15)
plt_s2_1.grid(True)
plt.plot( s_2_1_1[:2205] )


# sd.play( s , 44100 )
```

```python
fig, plots = plt.subplots(4,1,figsize=(12,12), sharex=True)
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

plots[0].grid(True)
plots[0].title.set_text(r"$\bf{s_{21} = s_{211}  + sin( 2π*f_2*t + π/4 )}$")
plots[0].plot( s_2_1_1[:2205], "b" )
plots[0].plot( s_2_2_1[:2205], "g" )
plots[0].plot( s_2_1[:2205], "k." )

plots[1].grid(True)
plots[1].title.set_text(r"$\bf{s_{22} = s_{211}  + sin( 2π*f_2*t + π/2 )}$")
plots[1].plot( s_2_1_1[:2205], "b" )
plots[1].plot( s_2_2_2[:2205], "g" )
plots[1].plot( s_2_2[:2205], "k." )

plots[2].grid(True)
plots[2].title.set_text(r"$\bf{s_{23} = s_{211}  + sin( 2π*f_2*t + π )}$")
plots[2].plot( s_2_1_1[:2205], "b" )
plots[2].plot( s_2_2_3[:2205], "g" )
plots[2].plot( s_2_3[:2205], "k." )

plots[3].grid(True)
plots[3].title.set_text(r"$\bf{s_{24} = s_{211}  + sin( 2π*f_2*t + 3π/2 )}$")
plots[3].plot( s_2_1_1[:2205], "b" )
plots[3].plot( s_2_2_4[:2205], "g" )
plots[3].plot( s_2_4[:2205], "k." )

# fig.suptitle(r"$\bf{Figure\ :\ s_2 = }$")

```

```python
sample_rate = 44100

sd.play( s_2_1, sample_rate )
sd.wait(4)
sd.play( s_2_2, sample_rate )
sd.wait(4)
sd.play( s_2_3, sample_rate )
sd.wait(4)
sd.play( s_2_4, sample_rate )
```

```python
fig, plots = plt.subplots(4,1,figsize=(12,12), sharex=True)
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

s1 = s_1 + s2_1
plots[0].grid(True)
plots[0].title.set_text(r"$\bf{s = s_1 + s_{21}}$")
plots[0].plot( s_1[:2205], "b" )
plots[0].plot( s_2_1[:2205], "g" )
plots[0].plot( s1[:2205], "k|" )

s2 = s_1 + s2_2
plots[1].grid(True)
plots[1].title.set_text(r"$\bf{s = s_1 + s_{22}}$")
plots[1].plot( s_1[:2205], "b" )
plots[1].plot( s_2_2[:2205], "g" )
plots[1].plot( s2[:2205], "k|" )

s3 = s_1 + s2_3
plots[2].grid(True)
plots[2].title.set_text(r"$\bf{s = s_1 + s_{23}}$")
plots[2].plot( s_1[:2205], "b" )
plots[2].plot( s_2_3[:2205], "g" )
plots[2].plot( s3[:2205], "k|" )

s4 = s_1 + s2_4
plots[3].grid(True)
plots[3].title.set_text(r"$\bf{s = s_1 + s_{24}}$")
plots[3].plot( s_1[:2205], "b" )
plots[3].plot( s2_4[:2205], "g" )
plots[3].plot( s4[:2205], "k|" )

fig.suptitle(r"$\bf{Figure\ :\ s = s_1 + s_2}$")
```
