```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import sounddevice as sd
```

<h1>/ a /</h1>

```python
sr_a, a = wf.read('./data/a_audio.wav')
# sd.play( a , sr_a )
duration_a = len(a[:,0])/sr_a
print("Sample rate: %d\nDuration: %.3fsec"%(sr_a, duration_a))
timesteps_per_msec = sr_a/1000
timesteps_100_msec = int(timesteps_per_msec*100)
print("Time steps per millisecond: %.1f\nTime steps per 100msec: %d"%(timesteps_per_msec, timesteps_100_msec))
middle_pos = int(len(a[:,0])/2)
# print(middle_pos)
fig = plt.figure(figsize=(15,3))
plt_a = fig.add_subplot(1, 1, 1)
plt_a.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude (db)")
plt.title(r"$\bf{Figure\ 1:\ Plot\ of\ the\ sound\ wave\ of\ phoneme\ /a/}$")
plt.gcf().subplots_adjust(bottom=0.15)
plt_a.plot( np.arange( a[middle_pos:middle_pos+timesteps_100_msec, 0].size ) , a[middle_pos:middle_pos+timesteps_100_msec, 0] )
time_period = 191
plt_a.plot( np.arange( a[middle_pos:middle_pos+time_period, 0].size ) , a[middle_pos:middle_pos+time_period, 0] )

a_signal = a[middle_pos:middle_pos+timesteps_100_msec, 0]
cycles_in_100msec = timesteps_100_msec/time_period

#Counted manually
# cycles_in_100msec = 23.2

freq = cycles_in_100msec*10
print("Cycles in 100msec: %.3f\nFrequency: %.2fHz ~ A#3/Bb3 (233.08Hz)"%(cycles_in_100msec,freq))

sd.play( a_signal , sr_a )
```

<h1>/ k /</h1>

```python
sr_k, k = wf.read('./data/k_audio.wav')
print(len(k[:,0]))
duration_k = len(k[:,0])/sr_k
print("Sample rate: %d\nDuration: %.3fsec"%(sr_k, duration_k))
timesteps_per_msec = sr_k/1000
timesteps_100_msec = int(timesteps_per_msec*100)
print("Time steps per millisecond: %.1f\nTime steps per 100msec: %d"%(timesteps_per_msec, timesteps_100_msec))

#I chose another timestep to start from, as a more representative sound of the consonant /k/
middle_pos = int(len(k[:,0])/2) - 6100
# print(middle_pos)
fig = plt.figure(figsize=(15,3))
plt_k = fig.add_subplot(1, 1, 1)
plt_k.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude (db)")
plt.title(r"$\bf{Figure\ 5b:\ Plot\ of\ the\ sound\ wave\ of\ phoneme\ /k/}$")
plt.gcf().subplots_adjust(bottom=0.15)
plt_k.plot( np.arange( k[middle_pos:middle_pos+timesteps_100_msec, 0].size ) , k[middle_pos:middle_pos+timesteps_100_msec, 0] )

k_signal = k[middle_pos:middle_pos+timesteps_100_msec, 0]

sd.play( k_signal , sr_k )
```

```python
# concat_signal = a_signal + s_signal + k_signal + o_signal
concat_signal_skao = np.hstack((s_signal,k_signal,a_signal,o_signal))
concat_signal_skao = concat_signal_skao.reshape(concat_signal_skao.shape[0],-1)

concat_signal_sako = np.hstack((s_signal,a_signal,k_signal,o_signal))
concat_signal_sako = concat_signal_sako.reshape(concat_signal_sako.shape[0],-1)

concat_signal_asko = np.hstack((a_signal,s_signal,k_signal,o_signal))
concat_signal_asko = concat_signal_asko.reshape(concat_signal_asko.shape[0],-1)

concat_signal_soka = np.hstack((s_signal,o_signal,k_signal,a_signal))
concat_signal_soka = concat_signal_soka.reshape(concat_signal_soka.shape[0],-1)

concat_signal_kosa = np.hstack((k_signal,o_signal,s_signal,a_signal))
concat_signal_kosa = concat_signal_kosa.reshape(concat_signal_kosa.shape[0],-1)

concat_signal_kaos = np.hstack((k_signal,a_signal,o_signal,s_signal))
concat_signal_kaos = concat_signal_kaos.reshape(concat_signal_kaos.shape[0],-1)

concat_signal_eskasa = np.hstack((e_signal,s_signal,k_signal,a_signal,s_signal,a_signal))
concat_signal_eskasa = concat_signal_eskasa.reshape(concat_signal_eskasa.shape[0],-1)

concat_signal_osa = np.hstack((o_signal,s_signal,a_signal))
concat_signal_osa = concat_signal_osa.reshape(concat_signal_osa.shape[0],-1)

concat_signal_ase = np.hstack((a_signal,s_signal,e_signal))
concat_signal_ase = concat_signal_ase.reshape(concat_signal_ase.shape[0],-1)

concat_signal_kaka = np.hstack((k_signal,a_signal,k_signal,a_signal))
concat_signal_kaka = concat_signal_kaka.reshape(concat_signal_kaka.shape[0],-1)

concat_signal_kakos = np.hstack((k_signal,a_signal,k_signal,o_signal,s_signal))
concat_signal_kakos = concat_signal_kakos.reshape(concat_signal_kakos.shape[0],-1)

concat_signal_se = np.hstack((s_signal,e_signal))
concat_signal_se = concat_signal_se.reshape(concat_signal_se.shape[0],-1)

concat_signal_esosa = np.hstack((e_signal,s_signal,o_signal,s_signal,a_signal))
concat_signal_esosa = concat_signal_esosa.reshape(concat_signal_esosa.shape[0],-1)

concat_signal_sasa = np.hstack((s_signal,a_signal,s_signal,a_signal))
concat_signal_sasa = concat_signal_sasa.reshape(concat_signal_sasa.shape[0],-1)


sr_original = 44100

sd.play( concat_signal_kakos , sr_original )
sd.wait(2)
sd.play( concat_signal_se , sr_original )
sd.wait(2)
sd.play( concat_signal_esosa , sr_original )
sd.wait(2)
sd.play( concat_signal_sasa , sr_original )
sd.wait(2)

wf.write("./results/audio_files/kakos.wav", sr_original, concat_signal_kakos)
wf.write("./results/audio_files/se.wav", sr_original, concat_signal_se)
wf.write("./results/audio_files/esosa.wav", sr_original, concat_signal_esosa)
wf.write("./results/audio_files/sasa.wav", sr_original, concat_signal_sasa)

new_duration = concat_signal_skao.shape[0] / sr_original
print("Duration of new mixed concatenated signal: %.2fsec"%(new_duration))

fig, plots = plt.subplots(4,1,figsize=(20,12), sharex=True)
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")

plots[0].grid(True)
plots[0].title.set_text("[ka'kos]")
plots[0].plot( np.arange( concat_signal_kakos[:,0].size ) , concat_signal_kakos[:,0] )

plots[1].grid(True)
plots[1].title.set_text("['se]")
plots[1].plot( np.arange( concat_signal_se[:,0].size ) , concat_signal_se[:,0] )

plots[2].grid(True)
plots[2].title.set_text("['esosa]")
plots[2].plot( np.arange( concat_signal_esosa[:,0].size ) , concat_signal_esosa[:,0] )

plots[3].grid(True)
plots[3].title.set_text("['sasa]")
plots[3].plot( np.arange( concat_signal_sasa[:,0].size ) , concat_signal_sasa[:,0] )


fig.suptitle(r"$\bf{Figure\ 6:\ Plot\ of\ the\ sound\ waves\ of\ phonemes\ /a/,\ /o/,\ /s/,\ /e/\ and\ /k/\ concatenated\ in\ multiple\ order}$")
```
