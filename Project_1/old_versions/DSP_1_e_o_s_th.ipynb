```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
```

 <h2>1. Phoneme /ε/ waveform ( greek [e] )</h2>

```python
sr_e, e = wf.read('./data/e_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_e = fig.add_subplot(1, 1, 1)
plt_e.grid(True)
TimePeriod = 178
# plt.plot( np.arange( a[1927:2633, 0].size ) , a[1927:2633, 0] )
plt_e.plot( np.arange( e[1927:1927+4*TimePeriod-6, 0].size ) , e[1927:1927+4*TimePeriod-6, 0] )
plt_e.plot( np.arange( e[1927:1927+TimePeriod, 0].size ) , e[1927:1927+TimePeriod, 0] )
# plt.savefig('./Figs/e_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( e[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( e[:, 0])))
```

<font size="3">
As the plot indicates, there is a relative/slight periodicity (visible pattern) in the soundwave, with a time period of about 178 time steps.
</font>

 <h2>2. Phoneme /o/ waveform ( greek [o] )</h2>

```python
sr_o, o = wf.read('./data/o_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_o = fig.add_subplot(1, 1, 1)
plt_o.grid(True)
TimePeriod = 203
plt_o.plot( np.arange( o[2000:2000+4*TimePeriod-8, 0].size ) , o[2000:2000+4*TimePeriod-8, 0])
plt_o.plot( np.arange( o[2000:2000+TimePeriod, 0].size ) , o[2000:2000+TimePeriod, 0])
# plt.savefig('./Figs/o_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( o[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( o[:, 0])))
```

<font size="3">
The pattern here, has a time period of about 203 time steps. Three peaks and three troughs were distinguished in the pattern.
</font>

<h2>Notes</h2><br>
<font size="3">
Both phonemes concern greek vowels ( [e], [o] ), the soundwaves of which exhibit patterns that are characterised by periodicity.
</font>

 <h2>3. Phoneme /s/ waveform</h2>

```python
sr_s, s = wf.read('./data/s_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_s = fig.add_subplot(1, 1, 1)
plt_s.grid(True)
plt_s.plot( np.arange( s[800:1550, 0].size ) , s[800:1550, 0] )
# plt.savefig('./Figs/s_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( s[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( s[:, 0])))
```

<font size="3">
The sound of "s" is more like an aperiodic hissing (noise) - with a possible random pattern in some parts, as observed from the plot above.  
</font>

 <h2>4. Phoneme /θ/ waveform</h2>

```python
sr_th, th = wf.read('./data/th_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_th = fig.add_subplot(1, 1, 1)
plt_th.grid(True)
plt_th.plot( np.arange( th[3000:3750, 0].size ) , th[3000:3750, 0])
# plt.savefig('./Figs/th_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( th[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( th[:, 0])))
```

<font size="3">
Here, the plot indicates random pattern occurrences. The waveform shows that the phoneme /θ/ is less intense than phoneme /s/ above. 
</font>

<h2>Notes</h2><br>
<font size="3">
Both phonemes concern voiceless consonants ( [s], [θ] ) with no visible consistent patterns in the produced soundwaves. 
</font>

```python
fig, plots = plt.subplots(4,1,figsize=(10,10), sharex=True)

steps_to_plot = 700

plots[0].grid(True)
plots[0].title.set_text('Phoneme / ε /')
plots[0].plot( np.arange( e[1927:1927+steps_to_plot, 0].size ) , e[1927:1927+steps_to_plot, 0] )

plots[1].grid(True)
plots[1].title.set_text('Phoneme / o /')
plots[1].plot( np.arange( o[2000:2000+steps_to_plot, 0].size ) , o[2000:2000+steps_to_plot, 0])

plots[2].grid(True)
plots[2].title.set_text('Phoneme / s /')
plots[2].plot( np.arange( s[800:800+steps_to_plot, 0].size ) , s[800:800+steps_to_plot, 0] )

plots[3].grid(True)
plots[3].title.set_text('Phoneme / θ /')
plots[3].plot( np.arange( th[3000:3000+steps_to_plot, 0].size ) , th[3000:3000+steps_to_plot, 0])

# plt.savefig('./Figs/all_wave.jpeg', dpi=3000)
```

<h2>Conclusions</h2>

<font size="3">
All the plots above are in the same zoom scale. By comparing the four representations visually, we can make some interesting conlusions about those two vowels ( [e], [o] ) and the two consonants ( [s], [θ] ):
    <ul>
        <li>The vowels’ sound waves have clear or almost clear patterns that repeat themselves throughout the signal in (almost) the same way. On the contrary, the consonants’ waves usually have random patterns without repetition consistency or even no patterns at all.
        <li>The consonants’ waveforms are more dense than the vowels’. They carry more information during the same time frame (e.g. 700 time steps above)
        <li>The above, might be a result of the the different articulation of each phoneme, as well as the nature of the sound produced by this specific movements of the vocal tract.
    </ul>
</font>
