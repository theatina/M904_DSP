```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
```

<font size="3">
Sound is the result of back and forth vibration of the particles of the medium through which the sound wave is moving. In the case of the air as the medium, the air particles are set in longitudinal motion, creating compressions and rarefactions.

In the project at hand, the sound waves of four greek phonemes are plotted, individually examined, analyzed and compared. The ‘x’ axis represents the time and the ‘y’ represents the attribute of amplitude.
The intensity of a sound at each time step (x-axis), in our case the phoneme’s, can be expressed via the amplitude value of the sound wave in db (decibel) at this particular time fragment (x-axis). The pitch depends on the frequency in which the vocal cords are vibrating, which is also depicted on the plot - the higher the frequency, the denser the signal (the more peaks and troughs in a time period / full cycle). A single frequency, constant in time, is called pure tone, while a sound consisting of multiple frequencies, is called complex tone. The vocal sounds are complex tones due to the nature of the production, in which multiple articulators contribute to the aforementioned procedure. Some phonemes’ audio signals are characterised as periodic ( / a / ) and others’ as aperiodic ( / s / ). 
</font>

 <h2>1. Phoneme /a/ waveform</h2> 

```python
sr_a, a = wf.read('./data/a_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_a = fig.add_subplot(1, 1, 1)
plt_a.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude (db)")
plt.title(r"$\bf{Figure\ 1:\ Plot\ of\ the\ sound\ wave\ of\ phoneme\ /a/}$")
plt.gcf().subplots_adjust(bottom=0.15)
TimePeriod = 189
# plt.plot( np.arange( a[:1000, 0].size ) , a[:1000, 0] )
plt_a.plot( np.arange( a[1897:1897+4*TimePeriod, 0].size ) , a[1897:1897+4*TimePeriod, 0] )
plt_a.plot( np.arange( a[1897:1897+TimePeriod, 0].size ) , a[1897:1897+TimePeriod, 0] )
# plt.savefig('./Figs/a_wave_part.jpeg', dpi=500)
print("Max Amplitude value throughout the wave: "+str(max( a[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( a[:, 0])))
```

<font size="3">
As Figure 1 indicates, some kind of periodicity (visible pattern) is detected in the sound wave, with a time period of about 189 time steps, where approximately 3 peaks and 3 troughs emerge in each. The vowel / a / is a central low and unrounded sonorous phoneme , suggesting that the vocal cords are semi-open while letting the air flow from the lungs towards the exit of the vocal tract. All the articulators are positioned in such a way to allow the unhindered air flow in the oral cavity.
</font>

 <h2>2. Phoneme /o/ waveform ( greek [o] )</h2>

```python
sr_o, o = wf.read('./data/o_audio.wav')
print(len(o[:,0]))
fig = plt.figure(figsize=(15,3))
plt_o = fig.add_subplot(1, 1, 1)
plt_o.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title(r"$\bf{Figure\ 2.1:\ Plot\ of\ the\ sound\ wave\ of\ phoneme\ /o/}$")
plt.gcf().subplots_adjust(bottom=0.15)
TimePeriod = 203
plt_o.plot( np.arange( o[2000:2000+4*TimePeriod-8, 0].size ) , o[2000:2000+4*TimePeriod-8, 0])
plt_o.plot( np.arange( o[2000:2000+TimePeriod, 0].size ) , o[2000:2000+TimePeriod, 0])
# plt.savefig('./Figs/o_wave_part.jpeg', dpi=500)
print("Max Amplitude value throughout the wave: "+str(max( o[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( o[:, 0])))
```

<font size="3">
The pattern here, has a time period of about 203 time steps. Three peaks and three troughs were distinguished in the pattern. The / o / vowel also resides in the category of the voiced phonemes, so the same apply to it (concerning the voicing). It is a mid-back rounded resonant vowel.
</font>

```python
fig = plt.figure(figsize=(15,3))
plt_ = fig.add_subplot(1, 1, 1)
plt_.grid(True)
colour1 = 'purple'
colour2 = 'orange'
blue_patch = mpatches.Patch(color=colour1, label='/ a / phoneme')
orange_patch = mpatches.Patch(color=colour2, label='/ o / phoneme')
plt.legend(handles=[blue_patch,orange_patch])
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title(r"$\bf{Figure\ 2.2:\ Plot\ of\ /a/\ and\ /o/\ phoneme’s\ respective\ sound\ waves}$")
plt.gcf().subplots_adjust(bottom=0.15)
steps_to_plot = 300
plt_.plot( np.arange( a[1897:1897+4*steps_to_plot, 0].size ) , a[1897:1897+4*steps_to_plot, 0], colour1 )
plt_.plot( np.arange( o[2000:2000+4*steps_to_plot-8, 0].size ) , o[2000:2000+4*steps_to_plot-8, 0], colour2)
# plt.savefig('./Figs/waves_a_o.jpeg', dpi=500)
```

<font size="3">
Both phonemes are vowels ( / a /, / o / ), the sound waves of which exhibit patterns that are characterised by periodicity. Generally, in the case of the vowels, the vocal cords are adducted and vibrating, finally producing an output sound that is a combination (sum) of the fundamental frequency and the harmonics of the vocal cords as well as the resonated sound waves (in the oral cavity) with characteristics depending on the rest of the articulators’ position and movements. The nature of the vocal sound wave as a complex tone is indicated by Figure 1 and  Figure 2, where each wave form seems to be the sum of multiple waves (e.g. harmonics), resulting in reduced smoothness and clarity of the pure fundamental tone. Phoneme /a/ is more sonorous than phoneme /o/, a remark made soon after a casual comparison between the amplitude attribute of two audio signals (Figure 2.2), which is reasonable, considering the higher constriction level of the sound from the articulators (e.g. rounded lips).
</font>

 <h2>3. Phoneme /s/ waveform</h2>

```python
sr_s, s = wf.read('./data/s_audio.wav')
print(len(s[:,0]))
fig = plt.figure(figsize=(15,3))
plt_s = fig.add_subplot(1, 1, 1)
plt_s.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title(r"$\bf{Figure\ 3:\ Plot\ of\ the\ sound\ wave\ of\ phoneme\ /s/}$")
plt.gcf().subplots_adjust(bottom=0.15)
plt_s.plot( np.arange( s[800:1600, 0].size ) , s[800:1600, 0] )
# plt.savefig('./Figs/s_wave_part.jpeg', dpi=500)
print("Max Amplitude value throughout the wave: "+str(max( s[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( s[:, 0])))
```

<font size="3">
The sound of fricative alveoral phoneme /s/ is more like an aperiodic hissing (white noise) with a possible random pattern in some parts of the signal, as observed from Figure 3. This sound is produced by air flowing through constricted articulators, producing audible frictional turbulence. The vocal cords are abducted, an articulator position deployed for the production of all the voiceless phonemes.
</font>

 <h2>4. Phoneme /g/ waveform</h2>

```python
sr_k, k = wf.read('./data/k_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_k = fig.add_subplot(1, 1, 1)
plt_k.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title(r"$\bf{Figure\ 4:\ Plot\ of\ the\ sound\ wave\ of\ phoneme\ /k/}$")
plt.gcf().subplots_adjust(bottom=0.15)
ymax = max( k[:, 0] )
ymin = min( k[:, 0] )
plt_k.plot( np.arange( k[1000:14000, 0].size ) , k[1000:14000, 0])
# plt_g.plot( np.arange( g[19000:30900, 0].size ) , g[19000:30900, 0])
# plt.savefig('./Figs/k_wave_part.jpeg', dpi=500)
print("Max Amplitude value throughout the wave: "+ str(ymax))
print("Min Amplitude value throughout the wave: "+ str(ymin))
```

<font size="3">
Here, the plot indicates an aperiodic sound without any patterns, commencing abruptly with a burst, followed by a stronger burst, where the maximum amplitude valude of the signal occurs, some time fragments afterwards. That signal suggests a stop phoneme, such as the voiceless velar plosive / k /. During the production of this consonant, the outgoing airstream is obstructed, then intraoral pressure is being built up and finally, as Figure 4 shows, the air is released, producing the noise burst which is the sound of / k /. 

Both phonemes concern voiceless consonants ( / s /, / k / ) with no consistent patterns in the produced sound waves. In the case of the purely voiceless sounds ( / s / ), there is no fundamental frequency ( no vibration of the vocal cords ), so the sound is considered white noise.

</font>

```python
fig, plots = plt.subplots(4,1,figsize=(10,10), sharex=True)
fig.text(0.05,0.5, "Amplitude", ha="center", va="center", rotation=90)
plt.xlabel("Time")
steps_to_plot = 4000

plots[0].grid(True)
plots[0].title.set_text('Phoneme / a /')
plots[0].plot( np.arange( a[1897:1897+steps_to_plot, 0].size ) , a[1897:1897+steps_to_plot, 0] )

plots[1].grid(True)
plots[1].title.set_text('Phoneme / o /')
plots[1].plot( np.arange( o[2000:2000+steps_to_plot, 0].size ) , o[2000:2000+steps_to_plot, 0])

plots[2].grid(True)
plots[2].title.set_text('Phoneme / s /')
plots[2].plot( np.arange( s[800:800+steps_to_plot, 0].size ) , s[800:800+steps_to_plot, 0] )

plots[3].grid(True)
plots[3].title.set_text('Phoneme / k /')
plots[3].plot( np.arange( k[3000:3000+steps_to_plot, 0].size ) , k[3000:3000+steps_to_plot, 0])
fig.suptitle("Figure 5: Plot of the sound waves of phonemes /a/, /o/, /s/ and /k/")


# plt.savefig('./Figs/all_waves_a_o_s_k.jpeg', dpi=500)
```

<h2>Conclusions</h2>

<font size="3">
All the plots above are in the same zoom scale ( time ). After comparison and examination of the four sound waves (Figure 5), we can make some interesting conclusions about the respective phonemes ( / a /, / o /, / s / and / k / ): 
    <ul>
        <li>The vowels’ sound waves have concrete or almost clear patterns that repeat themselves throughout the signal in (almost) the same way (periodic sound). On the contrary, the consonants’ waves usually have random patterns without repetition consistency or even no patterns at all (aperiodic sound). 
        <li>The consonants’ waveforms are more dense than the vowels’. They carry more information within the same time frame (4000 time steps), especially phoneme / s /. 
        <li>The above, might be a result of the different articulation of each phoneme, as well as the nature of the sound produced by the specific movements of the vocal tract. 
        <li>The sample rate of the four audio signals above is 44100Hz, which means if they were to be sampled with a bigger rate, such as 88.2kHz or even greater, we might have spotted more characteristics and differences, as the wave form would have been closer to the analogue (actual) signal.
    </ul>
