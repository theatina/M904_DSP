```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
```

 <h2>1. Phoneme /a/ waveform</h2> 

```python
sr_a, a = wf.read('./data/a_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_a = fig.add_subplot(1, 1, 1)
plt_a.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.gcf().subplots_adjust(bottom=0.15)
TimePeriod = 189
# plt.plot( np.arange( a[:1000, 0].size ) , a[:1000, 0] )
plt_a.plot( np.arange( a[1897:1897+4*TimePeriod, 0].size ) , a[1897:1897+4*TimePeriod, 0] )
plt_a.plot( np.arange( a[1897:1897+TimePeriod, 0].size ) , a[1897:1897+TimePeriod, 0] )
plt.savefig('./Figs/a_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( a[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( a[:, 0])))
```

<font size="3">
As the plot indicates, there is a relative/slight periodicity (visible pattern) in the soundwave, with a time period of about 178 time steps.
Front - low 
</font>

 <h2>2. Phoneme /o/ waveform ( greek [o] )</h2>

```python
sr_o, o = wf.read('./data/o_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_o = fig.add_subplot(1, 1, 1)
plt_o.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.gcf().subplots_adjust(bottom=0.15)
TimePeriod = 203
plt_o.plot( np.arange( o[2000:2000+4*TimePeriod-8, 0].size ) , o[2000:2000+4*TimePeriod-8, 0])
plt_o.plot( np.arange( o[2000:2000+TimePeriod, 0].size ) , o[2000:2000+TimePeriod, 0])
plt.savefig('./Figs/o_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( o[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( o[:, 0])))
```

<font size="3">
The pattern here, has a time period of about 203 time steps. Three peaks and three troughs were distinguished in the pattern.
</font>

<h2>Notes</h2><br>
<font size="3">
Both phonemes concern greek vowels ( [e], [o] ), the soundwaves of which exhibit patterns that are characterised by periodicity.
</font>

 <h2>3. Phoneme /s/ waveform</h2>

```python
sr_s, s = wf.read('./data/s_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_s = fig.add_subplot(1, 1, 1)
plt_s.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.gcf().subplots_adjust(bottom=0.15)
plt_s.plot( np.arange( s[800:1600, 0].size ) , s[800:1600, 0] )
plt.savefig('./Figs/s_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+str(max( s[:, 0])))
print("Min Amplitude value throughout the wave: "+str(min( s[:, 0])))
```

<font size="3">
The sound of "s" is more like an aperiodic hissing (noise) - with a possible random pattern in some parts, as observed from the plot above. This sound is produced by air flowing through constricted articulators, producing audible frictional turbulence, which is the phoneme /s/.
</font>

 <h2>4. Phoneme /g/ waveform</h2>

```python
sr_k, k = wf.read('./data/k_audio.wav')
fig = plt.figure(figsize=(15,3))
plt_k = fig.add_subplot(1, 1, 1)
plt_k.grid(True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.gcf().subplots_adjust(bottom=0.15)
ymax = max( k[:, 0] )
ymin = min( k[:, 0] )
plt_k.plot( np.arange( k[1000:14000, 0].size ) , k[1000:14000, 0])
# plt_g.plot( np.arange( g[19000:30900, 0].size ) , g[19000:30900, 0])
plt.savefig('./Figs/k_wave_part.jpeg', dpi=2000)
print("Max Amplitude value throughout the wave: "+ str(ymax))
print("Min Amplitude value throughout the wave: "+ str(ymin))
```

<font size="3">
Here, the plot indicates an aperiodic sound without any patterns, commencing abruptly with a burst, followed by a stronger burst some milliseconds afterwards. That signal suggests a stop phoneme, such as the voiceless /k/. During the production of this consonant, the outgoing airstream is obstructed, then intraoral pressure is being built up and finally, as the graph shows, the air is released producing the noise burst which is the sound of /k/.
</font>

<h2>Conclusions</h2>

<font size="3">
All the plots above are in the same zoom scale. By comparing the four representations visually, we can make some interesting conclusions about those two vowels ( [a], [o] ) and the two consonants ( [s], [k] ):
    <ul>
        <li>The vowels’ sound waves have clear or almost clear patterns that repeat themselves throughout the signal in (almost) the same way. On the contrary, the consonants’ waves usually have random patterns without repetition consistency or even no patterns at all.
        <li>The consonants’ waveforms are more dense than the vowels’. They carry more information during the same time frame (4000 time steps), especially phoneme /s/.
        <li>The above, might be a result of the the different articulation of each phoneme, as well as the nature of the sound produced by this specific movements of the vocal tract.
    </ul>
</font>
